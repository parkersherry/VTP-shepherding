```python
import pickle
import numpy as np
import matplotlib.pyplot as plt
```

```python
file_path = 'allDisplacementsList.pkl'

with open(file_path, 'rb') as file:
    
    norms = pickle.load(file)
```

```python
histNumber = 1
means = []
stds = []
counter = 0
for l in norms:
    counter+=1

    if counter == 11:
        continue
    arr = np.array(l)
    means.append(np.mean(arr))
    stds.append(np.std(arr))
    plt.hist(arr[arr<=5],bins=5)
    plt.xlabel("Pixels/Frame")
    plt.ylabel("Count")
    plt.title("Video #" + str(histNumber))
    plt.show() 

    #histNumber+=1
# plt.hist(means)
# plt.show()
# plt.hist(stds)
# plt.show()

x = np.linspace(1,14,14)
print(np.size(means))
plt.scatter(x,means,c='blue')
plt.scatter(x,stds,c='red')
plt.show()
```
